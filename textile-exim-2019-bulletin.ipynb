# Libraries

```python
import pandas as pd
import numpy as np
```

# Create DataFrame

```python
exim_2019 = pd.read_csv ('exim_2019.csv')

exim_2019
```

```python
hs_code = pd.read_excel('HS CODE.xlsx')
hs_code
```

# Data Preparation

```python
exim_2019 = exim_2019.merge(hs_code, how = 'right')
exim_2019 = exim_2019.fillna(0)

exim_2019
```

```python
exim_2019['USD'] = np.where(exim_2019['CIF VALUE (USD)'] == 0,
                            exim_2019['FOB VALUE (USD)'],
                            exim_2019['CIF VALUE (USD)'])

exim_2019['COUNTRY'] = np.where(exim_2019['COUNTRY OF DESTINATION'] == 0,
                                exim_2019['COUNTRY OF ORIGIN'],
                                exim_2019['COUNTRY OF DESTINATION'])

exim_2019
```

```python
exim_2019 = exim_2019.rename(columns = {'NET WEIGHT (KG)': 'KG'})

exim_2019.loc[exim_2019['FOB VALUE (USD)'] <= 0, 'EXIM'] = 'IMPORT' 
exim_2019.loc[exim_2019['FOB VALUE (USD)'] > 0, 'EXIM'] = 'EXPORT'

exim_2019
```

```python
column_names = ['YEAR', 'MONTH', 'EXIM', 'BULLETIN CODE', 'CATEGORY & COMMODITY',
                'DESCRIPTION', 'HS CODE', 'COUNTRY', 'KG', 'USD']

exim_2019 = exim_2019.reindex(columns = column_names)
exim_2019
```

```python
exim_2019 = exim_2019[(exim_2019.KG > 0) | (exim_2019.USD > 1)]

exim_2019 = exim_2019.sort_values('BULLETIN CODE')

exim_2019 = exim_2019.filter(['YEAR', 'MONTH', 'EXIM', 'CATEGORY & COMMODITY',
                              'BULLETIN CODE', 'HS CODE', 'COUNTRY', 'KG', 'USD'])

exim_2019
```

## Split DataFrame (1)

```python
textile = exim_2019[exim_2019['BULLETIN CODE'] <= 130]
textile = textile.filter(['YEAR', 'MONTH', 'EXIM', 'BULLETIN CODE', 'CATEGORY & COMMODITY', 'HS CODE', 'COUNTRY', 'KG', 'USD'])

raw = exim_2019[exim_2019['BULLETIN CODE'] >= 130]
raw = raw.filter(['YEAR', 'MONTH', 'EXIM', 'BULLETIN CODE', 'CATEGORY & COMMODITY', 'HS CODE', 'COUNTRY', 'KG', 'USD'])
```

# TEXTILE and RAW MATERIAL

```python
# Column Filter
textile1 = textile.filter(['YEAR', 'MONTH', 'EXIM', 'CATEGORY & COMMODITY', 'BULLETIN CODE', 'KG', 'USD'])

pv_textile_mt = textile1.pivot_table(index = ['YEAR', 'MONTH', 'BULLETIN CODE', 'CATEGORY & COMMODITY'],
                                    columns = 'EXIM',
                                    aggfunc = 'sum',
                                    fill_value = 0).swaplevel(axis = 1).sort_index(1)

# Stack
pv_textile_mt = pv_textile_mt.stack(level = -1)

pv_textile_mt
```

```python
pv_raw_mt = raw.pivot_table(index = ['YEAR', 'MONTH', 'BULLETIN CODE', 'HS CODE', 'CATEGORY & COMMODITY'],
                            columns = 'EXIM',
                            aggfunc = 'sum',
                            fill_value = 0).swaplevel(axis = 1).sort_index(1)

# Stack
pv_raw_mt = pv_raw_mt.stack(level = -1)

pv_raw_mt
```

## Cumulative

```python
pv_textile_cml = textile1.pivot_table(index = ['YEAR', 'BULLETIN CODE', 'CATEGORY & COMMODITY'],
                                      columns = 'EXIM',
                                      aggfunc = 'sum',
                                      fill_value = 0).swaplevel(axis = 1).sort_index(1)

# Stack
pv_textile_cml = pv_textile_cml.stack(level = -1)

pv_textile_cml
```

```python
pv_raw_cml = raw.pivot_table(index = ['YEAR', 'BULLETIN CODE', 'HS CODE', 'CATEGORY & COMMODITY'],
                             columns = 'EXIM',
                             aggfunc = 'sum',
                             fill_value = 0).swaplevel(axis = 1).sort_index(1)

# Stack
pv_raw_cml = pv_raw_cml.stack(level = -1)

pv_raw_cml
```

# Save to EXCEL (1)

```python
# Create a Pandas Excel writer using XlsxWriter as the engine.
writer = pd.ExcelWriter('TEXTILE and RAW.xlsx', engine = 'xlsxwriter')

# Write each DataFrame to a different worksheet.
pv_textile_mt.to_excel(writer, sheet_name = 'TEX_MONTH')
pv_raw_mt.to_excel(writer, sheet_name = 'RAW_MONTH')


pv_textile_cml.to_excel(writer, sheet_name = 'TEX_Cumulative')
pv_raw_cml.to_excel(writer, sheet_name = 'RAW_Cumulative')

# Close the Pandas Excel writer and output the Excel file.
writer.save()
```
